```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm_notebook
```

```python
base_url = "https://www.premiumbeautynews.com/"

def parse_url(url):
  response = requests.get(url)
  content = response.content
  parse_content = BeautifulSoup(content,'lxml')
  return parse_content


```

```python
def extract_any_postdata(post_url):
  soup_2 = parse_url(post_url)
  title = soup_2.find('h1',{'class':'article-title'}).text
  date_time = soup_2.find('div',{'class':"col-md-6 header-date"}).text
  abstract = soup_2.find('h2',{'class':'article-intro'}).text
  content = soup_2.find('div',{'class':'article-text'}).text

  data = {'title':title,'date_time':date_time,'abstract':abstract,'content':content,'url':post_url}
  return data
```

```python
url = "https://www.premiumbeautynews.com/en/markets-trends/"
soup = parse_url(url)
section = soup.find('section',{'class':'content'})
posts = section.findAll('div',{"class":"post-style1 col-md-6"})


```

```python

next_button = ''
posts_data = []
count = 1
while next_button is not None:
  for post in tqdm_notebook(posts,leave=False):
    uri = post.find('h4').find('a')['href']
    post_url = base_url+uri
    data = extract_any_postdata(post_url)
    posts_data.append(data)



  next_button = soup.find('span',{'class':'next'})
  if next_button is not None:
    post_url = base_url+next_button.find('a')['href']
    count +=1
    print(count)







```

```python
df = pd.DataFrame(posts_data)
print(len(df))
```
